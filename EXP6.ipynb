```python

'''6i. Salary Calculation with Allowances
       Calculate the gross salary of an employee using python and
       If basic <= 10000 : HRA = 20%, DA = 80%
       If basic <= 20000 : HRA = 25%, DA = 90%
       Otherwise: HRA = 30%, DA = 95%
       gross= basic + hra + da'''

basic=int(input("Enter Salary: "))
if basic <= 10000:
  hra=0.2*basic
  da=0.8*basic
elif basic <= 20000:
  hra=0.25*basic
  da=0.9*basic
else:
  hra=0.3*basic
  da=0.95*basic
print("Gross Salary",basic+hra+da)
```
